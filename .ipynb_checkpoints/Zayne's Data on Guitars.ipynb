```python
# Eliminate string characters from numeric data
data['Number of Ratings'] = data['Number of Ratings'].str.replace('No Ratings','0')
data['Rating'] = data['Rating'].str.replace('No Ratings','0')
                                            
data2['Number of Ratings'] = data2['Number of Ratings'].str.replace('No Ratings','0')
data2['Rating'] = data2['Rating'].str.replace('No Ratings','0')
                                            
data3['Number of Ratings'] = data3['Number of Ratings'].str.replace('No Ratings','0')
data3['Rating'] = data3['Rating'].str.replace('No Ratings','0')

data4['Number of Ratings'] = data4['Number of Ratings'].str.replace('No Ratings','0')
data4['Rating'] = data4['Rating'].str.replace('No Ratings','0')
```

```python
# Exclude redundant rows
NoDupsData = data.drop_duplicates()
NoDupsData2 = data2.drop_duplicates()
NoDupsData3 = data3.drop_duplicates()
```

###### 3. Merge dataframes from separate json files as appropriate

```python
# Create the inner merge of the two dataframes
dataInnerMerge = pd.merge(NoDupsData, NoDupsData2, \
                          on=['Description','Name','Price','Number of Ratings','Rating'],\
                          how='inner')
dataInnerMerge2= pd.merge(dataInnerMerge, NoDupsData3, \
                          on=['Description','Name','Price','Number of Ratings','Rating'],\
                          how='inner')
```
